```python
n_estimators = [100,500,900,1100]
max_depth = [2,3,5,10,15,29]
booster = ['gbtree','gblinear']
learning_rate = [0.05,0.1,0.15,0.18,0.20,0.22,0.28]
min_child_weight = [1,2,3,4,5,6,7,8,9]
hyperpara = {
    'n_estimators':n_estimators,
    'max_depth' : max_depth,
    'booster': booster,
    'learning_rate' : learning_rate,
    'min_child_weight' : min_child_weight
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
randoncv = RandomizedSearchCV(estimator=regressor,param_distributions=hyperpara,cv=10,n_iter=50,scoring='neg_mean_absolute_error',n_jobs=-1,verbose=5,
                             return_train_score=True,random_state=6)
```

```python
randoncv.fit(X_train,Y_train)
```

```python
randoncv.best_estimator_
```
